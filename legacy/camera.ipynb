```python
import cv2

# Acceder a la cámara (0 para la cámara interna, 1 o más para cámaras externas)
cap = cv2.VideoCapture(0)

# Comprobar si la cámara está abierta
if not cap.isOpened():
    print("No se pudo acceder a la cámara")
    exit()

# Capturar fotogramas en un bucle
while True:
    ret, frame = cap.read()
    if not ret:
        print("No se pudo recibir la imagen.")
        break

    # Mostrar el fotograma
    cv2.imshow('Frame', frame)

    # Presionar 'q' para salir del bucle
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
import cv2
from IPython.display import display, clear_output
import PIL.Image
import time

# Inicializar la cámara
cap = cv2.VideoCapture(0)

try:
    while True:
        # Leer un fotograma de la cámara
        ret, frame = cap.read()
        if not ret:
            print("No se pudo recibir la imagen.")
            break

        # Convertir a formato RGB para mostrar en Jupyter
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = PIL.Image.fromarray(rgb_frame)

        # Mostrar la imagen en Jupyter Notebook
        clear_output(wait=True)
        display(img)

        # Añadir un delay para actualizar la imagen cada 100ms
        time.sleep(0.1)
        
        # Aquí puedes añadir el código para el reconocimiento facial
        # Ejemplo: `result = your_face_recognition_model.predict(frame)`
        
except KeyboardInterrupt:
    # Permite detener el bucle con una interrupción manual
    print("Interrumpido por el usuario.")

finally:
    # Liberar la cámara y cerrar las ventanas
    cap.release()
    cv2.destroyAllWindows()
```
